```python
import pandas as pd
```

```python
df_exams = pd.read_csv('StudentsPerformance.csv')
df_exams.head()
```

# Ordenar DataFrame

```python
# ordenear por una columna
df_exams.sort_values(by = 'math score')
```

```python
# ordenar descendente por una columna
df_exams.sort_values(by = 'math score', ascending = False)
```

```python
# ordenar descendete por multiples columnas
df_exams.sort_values(['math score', 'reading score'], ascending = False) # crea copia
```

```python
# orden descendete por multiples columnas y update dataframe
df_exams.sort_values(['math score', 'reading score'], ascending = False, inplace = True)
```

```python
# orden descendete con una funcion key
df_exams.sort_values('race/ethnicity', key = lambda col : col.str.lower())
```
