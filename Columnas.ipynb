## 1.2 Agregar una nueva columna con un array

El dataframe tiene 1000 filas; por ello, tenemos que crear un array de 1000 elementos

```python
# importar numpy
import numpy as np
```

```python
# crear un array de 1000 elementos
language_score = np.arange(0, 1000)
```

```python
# tamaño del array
len(language_score)
```

```python
# agregar una nueva columna al dataframe con un array
df_exams['language score'] = language_score # sobreescribiendo
df_exams
```

```python
# crear numeros enteros aleatorios entre 1 y 100 (exclusivo)
language_score = np.random.randint(1, 100, size = 1000)
```

```python
# el minimo valor es inclusivo y el maximo valor exclusivo
min(language_score)
max(language_score)
```

```python
# agregar una nueva columna al dataframe con un array
df_exams['language score'] = language_score
df_exams
```
