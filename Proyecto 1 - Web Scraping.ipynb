```python
import pandas as pd
```

# Leer .csv de URL con Pandas
Website: [Football Data](https://www.football-data.co.uk/data.php)

```python
# leer 1 archivo csv de un website
df_premier26 = pd.read_csv('https://www.football-data.co.uk/mmz4281/2526/E0.csv')
```

```python
# mostrar dataframe
df_premier26
```

```python
# renombrar columnas
df_premier26.rename(columns = {'Date' : 'date',
                              'HomeTeam' : 'home_team',
                              'AwayTeam' : 'away_team',
                              'FTHG' : 'home_goals',
                              'FTAG' : 'away_goals'}, inplace = True)
```
